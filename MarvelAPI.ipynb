Importing libraries

```python
from marvel import Marvel #usage of the pip
import pandas as pd
```

Usage

```python
m = Marvel("7e674820cfa31156ce7edca15387401c","762964d88b102e42c8bb0f063cc368666f0cb22f") #m=Marvel("PUBLIC_KEY","PRIVATE_KEY")
```

Character object example

```python
characters = m.characters
```

```python
name = "Captain" #string/ character name that begin with the specified string 
```

```python
all_characters = m.characters.all(nameStartsWith=name) #nameStartsWith parameter
```

Let's print the variable `all_characters`to see all the info

```python
#all_characters #it does not looks pretty
```

Lets print the id character and the name character that starts with the variable `name`





```python
for i in range(len(all_characters["data"]["results"])):
  
  print(all_characters["data"]["results"][i]["id"], 
        all_characters["data"]["results"][i]["name"])
```

Data frame of the results data of the request `all_characters`

```python
df_characters = pd.DataFrame(all_characters["data"]["results"])
df_characters
```

Select a character Id to test `characters.stories()`function

```python
story = m.characters.stories(df_characters["id"][0]) #Id = 	1009220
df_story = pd.DataFrame(story["data"]["results"])
df_story
```

Lets print the stories titles wich the character `Id = 1009220` appears

```python
print(df_story["title"])
  
```

Now let's try the `characters.comics()` function

```python
comic = m.characters.comics(1009220)
df_comic = pd.DataFrame(comic["data"]["results"])
df_comic
```
